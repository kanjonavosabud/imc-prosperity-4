## OSMIUM

```python
osmium_prices_day_1_df = prices_day_1_df[prices_day_1_df['product'] == 'ASH_COATED_OSMIUM']
osmium_trades_day_1_df = trades_day_1_df[trades_day_1_df['symbol'] == 'ASH_COATED_OSMIUM']
osmium_prices_day_2_df = prices_day_2_df[prices_day_2_df['product'] == 'ASH_COATED_OSMIUM']
osmium_trades_day_2_df = trades_day_2_df[trades_day_2_df['symbol'] == 'ASH_COATED_OSMIUM']
osmium_prices_day_3_df = prices_day_3_df[prices_day_3_df['product'] == 'ASH_COATED_OSMIUM']
osmium_trades_day_3_df = trades_day_3_df[trades_day_3_df['symbol'] == 'ASH_COATED_OSMIUM']

osmium_prices_day_1_df
```

```python
plt.figure(figsize=(14, 6))
plt.plot(osmium_prices_day_1_df['timestamp'], osmium_prices_day_1_df['ask_price_1'], label='Best Ask Price', color='red', linewidth=1.5)
plt.plot(osmium_prices_day_1_df['timestamp'], osmium_prices_day_1_df['bid_price_1'], label='Best Bid Price', color='green', linewidth=1.5)

# Plot the bid-ask spread as a filled area
plt.fill_between(
    osmium_prices_day_1_df['timestamp'],
    osmium_prices_day_1_df['bid_price_1'],
    osmium_prices_day_1_df['ask_price_1'],
    color='orange',
    alpha=0.2,
    label='Bid-Ask Spread'
)

plt.title("Tomatoes Bid-Ask Spread Over Time (Day -2)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Combine OSMIUM Day 1, 2, and 3 price data into one dataframe, aligning timestamps for continuity.

# Make a shifted copy of day 2 to adjust its timestamps so it follows day 1
osmium_prices_day_2_df_shifted = osmium_prices_day_2_df.copy()
last_timestamp_day_1 = osmium_prices_day_1_df['timestamp'].iloc[-1]
osmium_prices_day_2_df_shifted['timestamp'] = osmium_prices_day_2_df_shifted['timestamp'] + last_timestamp_day_1 + 100

# Similarly, shift day 3 to follow day 2
osmium_prices_day_3_df_shifted = osmium_prices_day_3_df.copy()
last_timestamp_day_2 = osmium_prices_day_2_df_shifted['timestamp'].iloc[-1]
osmium_prices_day_3_df_shifted['timestamp'] = osmium_prices_day_3_df_shifted['timestamp'] + last_timestamp_day_2 + 100

# Concatenate all three shifted and original dataframes
osmium_combined_df = pd.concat(
    [osmium_prices_day_1_df, osmium_prices_day_2_df_shifted, osmium_prices_day_3_df_shifted],
    axis=0,
    ignore_index=True
)

osmium_combined_df
```

```python
plt.figure(figsize=(14, 6))
plt.plot(osmium_combined_df['timestamp'], osmium_combined_df['ask_price_1'], label='Best Ask Price', color='red', linewidth=1.5)
plt.plot(osmium_combined_df['timestamp'], osmium_combined_df['bid_price_1'], label='Best Bid Price', color='green', linewidth=1.5)

# Plot the bid-ask spread as a filled area
plt.fill_between(
    osmium_combined_df['timestamp'],
    osmium_combined_df['bid_price_1'],
    osmium_combined_df['ask_price_1'],
    color='orange',
    alpha=0.2,
    label='Bid-Ask Spread'
)

plt.title("OSMIUM Bid-Ask Spread Over Time (Day -2)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## PEPPER ROOT

```python
# 1. Prepare Pepper Root: Shift Day 2 and Day 3 so their timestamps continue after Day 1
pepper_prices_day_1_df = prices_day_1_df[prices_day_1_df['product'] == 'INTARIAN_PEPPER_ROOT'].copy()
pepper_prices_day_2_df = prices_day_2_df[prices_day_2_df['product'] == 'INTARIAN_PEPPER_ROOT'].copy()
pepper_prices_day_3_df = prices_day_3_df[prices_day_3_df['product'] == 'INTARIAN_PEPPER_ROOT'].copy()

pepper_prices_day_1_df
```

```python
# Plot Pepper Root Bid-Ask Spread Over Time
plt.figure(figsize=(14, 6))
plt.plot(pepper_prices_day_1_df['timestamp'], pepper_prices_day_1_df['ask_price_1'], label='Best Ask Price', color='red', linewidth=1.5)
plt.plot(pepper_prices_day_1_df['timestamp'], pepper_prices_day_1_df['bid_price_1'], label='Best Bid Price', color='green', linewidth=1.5)

# Plot the bid-ask spread as a filled area
plt.fill_between(
    pepper_prices_day_1_df['timestamp'],
    pepper_prices_day_1_df['bid_price_1'],
    pepper_prices_day_1_df['ask_price_1'],
    color='orange',
    alpha=0.2,
    label='Bid-Ask Spread'
)

plt.title("Pepper Root (INTARIAN_PEPPER_ROOT) Bid-Ask Spread Over Time (Day -2)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Shift day 2 to follow day 1
pepper_prices_day_2_df_shifted = pepper_prices_day_2_df.copy()
last_timestamp_day_1_pepper = pepper_prices_day_1_df['timestamp'].iloc[-1]
pepper_prices_day_2_df_shifted['timestamp'] = pepper_prices_day_2_df_shifted['timestamp'] + last_timestamp_day_1_pepper + 100

# Shift day 3 to follow day 2 (already shifted)
pepper_prices_day_3_df_shifted = pepper_prices_day_3_df.copy()
last_timestamp_day_2_pepper = pepper_prices_day_2_df_shifted['timestamp'].iloc[-1]
pepper_prices_day_3_df_shifted['timestamp'] = pepper_prices_day_3_df_shifted['timestamp'] + last_timestamp_day_2_pepper + 100

# Concatenate all three shifted and original dataframes for Pepper Root
pepper_combined_df = pd.concat(
    [pepper_prices_day_1_df, pepper_prices_day_2_df_shifted, pepper_prices_day_3_df_shifted],
    axis=0,
    ignore_index=True
)

pepper_combined_df

# Plot Pepper Root Bid-Ask Spread Over Time
plt.figure(figsize=(14, 6))
plt.plot(pepper_combined_df['timestamp'], pepper_combined_df['ask_price_1'], label='Best Ask Price', color='red', linewidth=1.5)
plt.plot(pepper_combined_df['timestamp'], pepper_combined_df['bid_price_1'], label='Best Bid Price', color='green', linewidth=1.5)

# Plot the bid-ask spread as a filled area
plt.fill_between(
    pepper_combined_df['timestamp'],
    pepper_combined_df['bid_price_1'],
    pepper_combined_df['ask_price_1'],
    color='orange',
    alpha=0.2,
    label='Bid-Ask Spread'
)

plt.title("Pepper Root (INTARIAN_PEPPER_ROOT) Bid-Ask Spread Over Time (combined)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Analysis of Bid and Ask Volumes for Pepper Root

# Let's look at the evolution of best bid and ask volumes over time
plt.figure(figsize=(14, 6))
plt.plot(
    pepper_combined_df['timestamp'],
    pepper_combined_df['bid_volume_1'],
    label='Best Bid Volume',
    color='blue',
    alpha=0.7
)
plt.plot(
    pepper_combined_df['timestamp'],
    pepper_combined_df['ask_volume_1'],
    label='Best Ask Volume',
    color='purple',
    alpha=0.7
)
plt.title("Pepper Root (INTARIAN_PEPPER_ROOT) Best Bid/Ask Volumes Over Time (combined)")
plt.xlabel("Timestamp")
plt.ylabel("Volume")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Statistical summary
bid_vol_mean = pepper_combined_df['bid_volume_1'].mean()
ask_vol_mean = pepper_combined_df['ask_volume_1'].mean()
bid_vol_std = pepper_combined_df['bid_volume_1'].std()
ask_vol_std = pepper_combined_df['ask_volume_1'].std()

print("Pepper Root - Best Bid Volume Mean:", bid_vol_mean)
print("Pepper Root - Best Ask Volume Mean:", ask_vol_mean)
print("Pepper Root - Best Bid Volume Std Dev:", bid_vol_std)
print("Pepper Root - Best Ask Volume Std Dev:", ask_vol_std)

# Visualize the distribution of bid and ask volumes
plt.figure(figsize=(12,5))
plt.hist(
    pepper_combined_df['bid_volume_1'].dropna(),
    bins=30, alpha=0.6, label='Bid Volume', color='blue'
)
plt.hist(
    pepper_combined_df['ask_volume_1'].dropna(),
    bins=30, alpha=0.6, label='Ask Volume', color='purple'
)
plt.title("Distribution of Best Bid and Ask Volumes (Pepper Root)")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
